## Método de la transformada inversa

A partir de $U\sim\operatorname{U}(0,1)$ queremos generar $X$ con función de distribución $F_X$, donde asumimos que $F_X$ es una función estrictamente creciente.

Queremos ver si podemos hallar una transformación estrictamente monótona $T:[0,1]\to\mathbb{R}$ tal que $T(U)\overset{d}{=}X$ entonces tendremos

\begin{align}
    F_X(x)
    &=\operatorname{P}[X\leq x] \\
    &=\operatorname{P}[T(U)\leq x] \\
    &=\operatorname{P}[U\leq T^{-1}(x)] \\
    &=T^{-1}(x)
\end{align}

para $x\in\mathbb{R}$ donde en el último paso se utilizó que $\operatorname{P}[U\leq y]=y$ cuando $U$ es uniforme en $(0,1)$.

https://es.wikipedia.org/wiki/M%C3%A9todo_de_la_transformada_inversa



```python
from scipy.stats import uniform, expon
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Generate exponential distributed random variables given the mean 
### and number of random variables
def exponential_inverse_trans(n=1,mean=1):
    U=uniform.rvs(size=n)
    X=-mean*np.log(1-U)
    actual=expon.rvs(size=n,scale=mean)
    
    plt.figure(figsize=(8,5))
    plt.hist(X, bins=50, alpha=0.5, label="Generated r.v.")
    plt.hist(actual, bins=50, alpha=0.5, label="Actual r.v.")
    plt.title("Generated vs Actual %i Exponential Random Variables" %n)
    plt.legend()
    plt.show()
    return X
```

```python
cont_example1=exponential_inverse_trans(n=100,mean=4)
cont_example2=exponential_inverse_trans(n=500,mean=4)
cont_example3=exponential_inverse_trans(n=1000,mean=4)
```
